```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.read_csv('netflix.csv')
df.info()
```

```python
df.head()
```

Ques1. In which year highest number of movies and web series are released ?

```python
year=df['release_year'].value_counts().head(10)
print(year)
year.plot(kind='bar',color='k')
plt.title('Most movies released in whcih year ?')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

Ques 2.which country has the highest number of films and web series production?

```python
country=df["country"].value_counts().head(10)
print(country) 
country.plot(kind="bar",color="purple")
plt.title("Top 10 Country in film\webseries production")
plt.xlabel("country")
plt.ylabel("count")
```

Ques 3. How many movies are there and Web Series are there on Netflix? 

```python
m_TV=df["type"].value_counts()
print(m_TV)
m_TV.plot(kind="pie",color="orange")
```

Ques 4. Top 10 genre on which highest numbers of movies are based on ?

```python
genre=df["listed_in"].value_counts().head(10)
print(genre)
genre.plot(kind="bar",color="green")
plt.title("Top 10 Genre")
plt.xlabel("Genre")
plt.ylabel("No. of movies/TV series")
plt.show()
```

Ques 5. Date on which most of the movies and webseries and released.

```python
date=df["date_added"].value_counts().head(10)
print(date)
date.plot(kind="bar",color="red")
plt.title("top 10 dates on which most of movies/TV series were released")
plt.xlabel("Date")
plt.ylabel("No. of movies/TV series")
```

Ques 6. Directors Who worked on the Highest number of movies or web series.

```python
director=df["director"].value_counts().head(10)
print(director)
director.plot(kind="bar",color="cyan")
plt.title("Top 10 Directors")
plt.xlabel("Director")
plt.ylabel("No. of movies/TV series")
plt.show()
```

Ques 7. Most repeated cast on netflix

```python
cas=df["cast"].value_counts().head(10)
print(cas)
cas.plot(kind="barh" ,color="magenta")
plt.title("Top 10 Cast")
plt.ylabel("Cast")
plt.xlabel("No. of movies/TV series")
```

Ques 8. Most movies and web series get Rated by which rating awards ?

```python
rating=df['rating'].value_counts().head(10)
print(rating)
rating.plot(kind='bar',color='brown')
plt.title('Top 10 Ratings')
plt.xlabel('Rating')
plt.ylabel('No. of movies/TV series')
```
